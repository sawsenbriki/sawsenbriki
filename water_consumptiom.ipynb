```python
water['Year']=water['Year'].fillna(water['Year'].mean())
```

```python
# Remove the trailing space from the column name
water['Industrial Water Use (%)'] = water['Industrial Water Use (%)'].fillna(water['Industrial Water Use (%)'].median())
```

```python
water.isnull().sum()
```

```python
# Replace NaN values with the median value of the same column
# median is more robust to outliers than mean
water['Per Capita Water Use (Liters per Day)'] = water['Per Capita Water Use (Liters per Day)'].fillna(water['Per Capita Water Use (Liters per Day)'].median())
```

```python
water.isnull().sum()
```

```python
water['Agricultural Water Use (%)']=water['Agricultural Water Use (%)'].fillna(water['Agricultural Water Use (%)'].mean())
```

```python
water.isnull().sum()
```

```python
water['ATotal Water Consumption (Billion Cubic Meters)']=water['Agricultural Water Use (%)'].fillna(water['Total Water Consumption (Billion Cubic Meters)'].mean())
```

```python
water['Household Water Use (%)']=water['Household Water Use (%)'].fillna(water['Household Water Use (%))'].mean())
```

```python
# Choisir un pays
pays = input("Entrez un pays : ")

# Filtrer
if pays not in water['Country'].unique():
    print("Ce pays n'existe pas.")
else:
    df_pays = water[water['Country'] == pays].sort_values(by='Year')
    # Créer un histogramme vertical
    plt.figure(figsize=(12, 6))
    sns.barplot(
        x='Year',
        y='Per Capita Water Use (Liters per Day)',
        data=df_pays,
        color='skyblue'
    )

    plt.title(f"💧 Consommation d’eau par habitant - {pays}", fontsize=16)
    plt.xlabel("Année")
    plt.ylabel("Litres par jour")
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.5)
    plt.tight_layout()
    plt.show()

    # Calculer l’évolution , this block was outside the else block, leading to IndentationError
    # Now correctly indented within the else block
    start = df_pays['Per Capita Water Use (Liters per Day)'].iloc[0]
    end = df_pays['Per Capita Water Use (Liters per Day)'].iloc[-1]
    diff = end - start

    print(f"\n➡️ En {df_pays['Year'].iloc[0]} : {start:.2f} litres/personne/jour")
    print(f"➡️ En {df_pays['Year'].iloc[-1]} : {end:.2f} litres/personne/jour")

    if diff > 0:
        print(f"\n📈 La consommation a **augmenté** de {diff:.2f} litres → possible gaspillage 💦")
    elif diff < 0:
        print(f"\n📉 La consommation a **diminué** de {abs(diff):.2f} litres → économie d’eau 🌱")
    else:
        print("\n➖ La consommation est **stable**.")

    # Afficher un graphe
    plt.figure(figsize=(10, 5))
    sns.barplot(x='Year', y='Per Capita Water Use (Liters per Day)', data=df_pays, palette='Blues')
    plt.title(f"Consommation d’eau par habitant (histogramme) - {pays}")
    plt.xlabel("Année")
    plt.ylabel("Litres par jour")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.grid(True)
    plt.show()
```

```python
# Supprimer les lignes avec valeurs manquantes dans les colonnes nécessaires
water = water.dropna(subset=[
    'Agricultural Water Use (%)',
    'Industrial Water Use (%)',
    'Household Water Use (%)'
])

# Calcul de la moyenne par pays pour chaque secteur
moyennes = water.groupby('Country')[[
    'Agricultural Water Use (%)',
    'Industrial Water Use (%)',
    'Household Water Use (%)'
]].mean()

# Trouver les tops
top_agricole = moyennes['Agricultural Water Use (%)'].idxmax()
top_industriel = moyennes['Industrial Water Use (%)'].idxmax()
top_domestique = moyennes['Household Water Use (%)'].idxmax()

# Afficher les meilleurs pays
print("🔍 Pays avec la plus grande consommation moyenne par secteur :")
print(f"🌾 Agriculture : {top_agricole} → {moyennes.loc[top_agricole, 'Agricultural Water Use (%)']:.2f}%")
print(f"🏭 Industrie : {top_industriel} → {moyennes.loc[top_industriel, 'Industrial Water Use (%)']:.2f}%")
print(f"🏠 Ménages : {top_domestique} → {moyennes.loc[top_domestique, 'Household Water Use (%)']:.2f}%")

# 🎨 Visualisation : barplot des moyennes par secteur
plt.figure(figsize=(12, 6))
moyennes_sorted = moyennes.sort_values(by='Agricultural Water Use (%)', ascending=False).head(10)
moyennes_sorted[['Agricultural Water Use (%)', 'Industrial Water Use (%)', 'Household Water Use (%)']].plot(
    kind='bar',
    figsize=(14, 6),
    colormap='Set2'
)

plt.title('🌍 Top 10 pays par consommation moyenne d’eau (par secteur)')
plt.ylabel('Utilisation moyenne (%)')
plt.xlabel('Pays')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.legend(title='Secteur')
plt.show()
```
